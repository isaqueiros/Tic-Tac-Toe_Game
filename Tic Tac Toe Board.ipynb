```python
from IPython.display import clear_output

def displayGame(board):
    clear_output()
    
    print(board[1]+'|'+board[2]+'|'+board[3])
    print('-----')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('-----')
    print(board[7]+'|'+board[8]+'|'+board[9])
```

```python
def playerType():
    player2 = ''
    player1 = input('\nPlayer 1: You would like to be X or O? ').upper()

    while not (player1 == 'X' or player1 == 'O'):
        print('Sorry, invalid answer. Type X or O.')
        player1 = input('Player 1: You would like to be X or O? ').upper()
    
    if player1== 'X':
        return ('X','O')
    elif player1 =='O':
        return ('O','X')
```

```python
def playerMove():
    indexes = [1,2,3]
    global tries
    rowChoice = int(input('Choose your play! \n Which row? (1 to 3): '))
    
    while rowChoice not in indexes:
        rowChoice = input('That is not a valid row! Choose a row from 1 to 3: ')
    row = int(rowChoice)
    
    positionChoice = int(input(f'Which space of row {row}? '))
    
    while positionChoice not in indexes:
        positionChoice = input(f'Invalid Choice! Which space of row {row}? (Choose between 1 to 3) ')
    position = int(positionChoice)
    
    
    if (row == 1 or position == 3):
        result = row*position
    elif (row==2 and position==1): result = 4
    elif (row==2 and position==2): result = 5
    elif (row==3 and position==1): result = 7
    elif (row==3 and position==2): result = 8
    
    while result in tries:
        result = int(input('This position is taken. Choose a position from 1 to 9: '))
        while result not in range(1,10):
            result = int(input('That is not a valid position! Choose a row from 1 to 9: '))
        
    tries.append(result)
    
    return result
```

```python
def playerMark(position, marker):
    global gameBoard
    gameBoard[position]=marker
```

```python
def board_Full(gameBoard):
    global boardFull
    
    if ' ' not in gameBoard:
        boardFull = True
    
    return boardFull
```

```python
def check_Win(gameBoard):
    
    global checkWin
    
    if (gameBoard[1:4]==['X','X','X'] or gameBoard[1:4]==['O','O','O']):
        checkWin = True
    elif (gameBoard[4:7]==['X','X','X'] or gameBoard[4:7]==['O','O','O']):
        checkWin = True
    elif (gameBoard[7:10]==['X','X','X'] or gameBoard[7:10]==['O','O','O']):
        checkWin = True
    elif (gameBoard[1:10:3]==['X','X','X'] or gameBoard[1:10:3]==['O','O','O']):
        checkWin = True
    elif (gameBoard[2:10:3]==['X','X','X'] or gameBoard[2:10:3]==['O','O','O']):
        checkWin = True
    elif (gameBoard[3:10:3]==['X','X','X'] or gameBoard[3:10:3]==['O','O','O']):
        checkWin = True
        
    return checkWin
```

```python
def restart():
    return input('Would ypu like to play again? ').lower().startswith('y')
```

```python
#The game:

gameBoard = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

print('Welcome to the Tic Tac Toe Game!\n')
playStart = input('Are you ready to Play? (Yes/No) ').lower()
checkWin = False
boardFull = False

if playStart.startswith('y'):
    gameOn = True
    displayGame(gameBoard)
    
    Player = 'Player 1'
    player1Mark, player2Mark = playerType()
    
else:
    gameOn = False

tries=[]

while gameOn == True:
    
    #Player 1 Move:
    
    if Player == 'Player 1':

        newMove = playerMove()

        playerMark(newMove, player1Mark)
        check_Win(gameBoard)
        board_Full(gameBoard)

        if (checkWin == True or boardFull == True):
            gameOn = False
            displayGame(gameBoard)
            if checkWin == True:
                print('Congrats! You won!')
            elif boardFull == True:
                print('Too bad, board full! No one won this time.')
            if not restart():
                break
        else:
            gameOn = True
        
        Player = 'Player 2'
        displayGame(gameBoard)
    
    #Player 2 Move:
    
    elif Player == 'Player 2':
        
        newMove = playerMove()

        playerMark(newMove, player2Mark)
        check_Win(gameBoard)
        board_Full(gameBoard)
        
        if (checkWin == True or boardFull == True):
            gameOn = False
            displayGame(gameBoard)
            if checkWin == True:
                print('Congrats! You won!')
            elif boardFull == True:
                print('Too bad, board full! No one won this time.')
        else:
            gameOn = True
        
        Player = 'Player 1'
        displayGame(gameBoard)
```
